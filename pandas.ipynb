```python
df2 = pd.DataFrame({
    "name": ['Alfred', 'Batman', 'Catwoman'],
    "toy": [np.nan, 'Batmobile', 'Bullwhip'],
    "born": [pd.NaT, pd.Timestamp("1940-04-25"), pd.NaT]
})
```

```python
df2.head()
```

**Droping na**

```python
df2.dropna() # it will drop row if there is atleast one na
```

```python
df2.dropna(how='all') # it will only drop row iff all are na
```

**Drop duplicates**

```python
df2.drop_duplicates(subset=['name'], keep=False)
```

**Other methods**

```python
df2.info()
```

```python
df2['toy'].value_counts(dropna=True)
```
